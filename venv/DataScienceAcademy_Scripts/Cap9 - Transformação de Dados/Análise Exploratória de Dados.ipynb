### Análise Exploratória de Dados

#### Neste notebook usaremos uma pesquisa recente nos EUA sobre o mercado de trabalho para programadores de software, visando visualizar uma investigação dos dados a fim de detectar problemas com os dados, necessidades de mais variáveis, falhas na organização e necessidades de transformação.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import colorsys
plt.style.use('seaborn-talk')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Carregando o Dataset de Dados-Pesquisas.csv
# coluna separada por ','
df = pd.read_csv('Dados-Pesquisa.csv', sep = ',', low_memory = False)
```

```python
# Lista todas as colunas
list(df)
```

### Distribuição da Idade

```python
# Gerando um histograma
# Age - coluna hist - histograma
df.Age.hist(bins = 60) # bins indica a quantidade de colunas do histograma
plt.xlabel('Idade')
plt.ylabel('Número de profissionais')
plt.title('Distribuição de Idade')
plt.show()
# A maioria dos profissionais que trabalham com programação de Software estão na faixa de idade
# Entre 20 e 30 anos, sendo 25 anos a idade mais frequente
```

### Distribuição de Sexo

```python
labels = df.Gender.value_counts().index
# Contabilizando os elementos por cada um dos labels
num = len(df.EmploymentField.value_counts().index)
```

```python
# Criando a lista de cores
listaHSV = [(x*1.0/num, 0.5, 0.5) for x in range(num)]
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))
```

```python
# Gráfico de Pizza
# fatias representa as fatias de pizza
fatias, texto = plt.pie(df.Gender.value_counts(), colors = listaRGB, startangle = 45)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(fatias, labels, bbox_to_anchor = (1.05, 1))
plt.title('Gênero')
plt.show()
# Conclui-se que grande parte dos profissionais na área de desenvolvimento são do sexo masculino
```
