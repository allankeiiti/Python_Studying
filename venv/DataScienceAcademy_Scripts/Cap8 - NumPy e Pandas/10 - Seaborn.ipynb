```python
import seaborn as sea
```

```python
df = sea.load_dataset('tips')
```

```python
df.head()
```

```python
df
```

```python
# O mÃ©todo joinplot cria plot de 2 variÃ¡veis com grÃ¡ficos bivariados e univariados
sea.jointplot('total_bill', 'tip', df, kind = 'reg');
```

```python
# O mÃ©todo lmplot() cria um plot com dados e modelos de regressÃ£o
sea.lmplot('total_bill', 'tip', df, col = 'smoker')
# Compara os TIPS de quem Ã© ou nÃ£o fumante
```

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
dfPandas = pd.DataFrame()
# Criando dataframe vazio
```

```python
# Alimentando o dataframe vazio
dfPandas['a'] = random.sample(range(1, 100), 25)
dfPandas['b'] = random.sample(range(1, 100), 25)
```

```python
dfPandas.head()
```

```python
dfPandas
```

```python
# Scatter Plot
sea.lmplot('a', 'b', data = dfPandas, fit_reg = True);
```

```python
# Density Plot
sea.kdeplot(dfPandas.b);
```

```python
sea.kdeplot(dfPandas.b, dfPandas.a)
```

```python
sea.distplot(dfPandas.a);
```

```python
# Histograma
plt.hist(dfPandas.a, alpha = .3)
sea.rugplot(dfPandas.a)
```

```python
# Box Plot
sea.boxplot([dfPandas.b, dfPandas.a]);
```

```python
# Violin Plot
sea.violinplot([dfPandas.a, dfPandas.b]);
```

```python
# HeatMap
sea.heatmap([dfPandas.b, dfPandas.a], annot = True, fmt = 'd');
```

```python
# Clustermap
sea.clustermap(dfPandas);
# Cluster Ã© um grupo de dados similares
```
