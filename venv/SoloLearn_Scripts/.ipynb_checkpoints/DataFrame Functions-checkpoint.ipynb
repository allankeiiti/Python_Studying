```python
import pandas as pd
import numpy as np
```

```python
data = {'Estado': ['Santa Catarina', 'Paraná', 'Goiás', 'Bahia', 'Minas Gerais'
                  ,'Minas Gerais', 'Paraná', 'Goiás'],
        'Ano': [2002, 2003, 2004, 2005, 2006
               ,2007, 2008, 2009],
        'Populacao': [1.5, 1.7, 3.6, 2.4, 2.9
                     ,3.0, 2.8, 3.0]}
data2 = {'Sigla': ['SC', 'PR', 'GO', 'BA', 'MG'
                  ,'MG', 'PA', 'GO'],
        'Ano': [2010, 2023, 2024, 2025, 2026
               ,2027, 2028, 2029],
        'Populacao': [2.5, 2.7, 4.6, 3.4, 3.9
                     ,4.0, 3.8, 4.0]}
```

```python
df1 = pd.DataFrame(data)
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
dfmerge = pd.merge(df1, df2, on=[df1['Estado'], df2['Sigla']], how = 'inner')
```

```python
dfgroup = dfmerge.groupby('key_1', as_index = False).max()
```
