### Análise Exploratória de Dados

#### Neste notebook usaremos uma pesquisa recente nos EUA sobre o mercado de trabalho para programadores de software, visando visualizar uma investigação dos dados a fim de detectar problemas com os dados, necessidades de mais variáveis, falhas na organização e necessidades de transformação.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import colorsys
plt.style.use('seaborn-talk')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# Carregando o Dataset de Dados-Pesquisas.csv
# coluna separada por ','
df = pd.read_csv('Dados-Pesquisa.csv', sep = ',', low_memory = False)
```

### Distribuição de Sexo

```python
labels = df.Gender.value_counts().index
# Contabilizando os elementos por cada um dos labels
num = len(df.EmploymentField.value_counts().index)
```

```python
# Criando a lista de cores
listaHSV = [(x*1.0/num, 0.5, 0.5) for x in range(num)]
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))
```

```python
# Gráfico de Pizza
# fatias representa as fatias de pizza
fatias, texto = plt.pie(df.Gender.value_counts(), colors = listaRGB, startangle = 45)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(fatias, labels, bbox_to_anchor = (1.05, 1))
plt.title('Gênero')
plt.show()
# Conclui-se que grande parte dos profissionais na área de desenvolvimento são do sexo masculino
```

### Idade X Horas de Aprendizagem

```python
import warnings
warnings.filterwarnings('ignore')

# Criando subset de dados
df9 = df.copy()
# Removendo valores NaN
df9 = df9.dropna(subset = ['HoursLearning'])
df9 = df9[df['Age'].isin(range(0, 70))]

# Definindo valores de X e Y
x = df9.Age
y = df9.HoursLearning

# Computando os valores e gerando o gráfico de linha de regressão
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, '.') # Se eu colocar '.', as variáveis serão representadas em Dots
plt.plot(x, m*x + b, '-', color = "red")
plt.xlabel('Idade')
plt.ylabel('Horas')
plt.title('idade por Horas de Treinamento')
plt.show()

# Buscando a relação entre idade e horas de treinamento
# A idade dos profissionais não afeta na quantidade de tempo gasto com treinamento
```

### Investimento em Capacitação X Expectativa Salarial

```python
# Qual a relação entre investimento de capacitação e Expectativa salarial dos colaboradores?

# Criando subset de dados
df10 = df.copy()
df10 = df10.dropna(subset = ['ExpectedEarning'])
df10 = df10[df['MoneyForLearning'].isin(range(0, 60000))]

# Definindo os valores de X e Y - X sendo a expectativa e Y o investimento em capacitação
x = df10.ExpectedEarning
y = df10.MoneyForLearning

# Computando os valores e gerando o gráfico
m, b = np.polyfit(x, y, 1)
plt.plot(x, y, '.')
plt.plot(x, m*x + b, '-', color = 'red')
plt.xlabel('Investimento em Capacitação')
plt.ylabel('Expectativa Salarial')
plt.title('Investimento em Capacitação X Expectativa Salarial')
plt.show()

# Os profissionais que investem tempo e dinheiro em capacitação e treinamento, em geral
# conseguem salários mais altos, embora alguns esperem salários altos, investimento 0 em treinamento
```
