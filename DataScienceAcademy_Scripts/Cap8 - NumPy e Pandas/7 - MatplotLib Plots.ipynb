## Construindo Plots

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.__version__
```

```python
# plt.plot (X, Y)
plt.plot([1, 3, 5], [2, 5, 7])
plt.show()
```

```python
x = [1, 4, 5]
y = [3, 7, 2]
```

```python
plt.plot(x, y)
plt.xlabel('Var 1')
plt.ylabel('Var 2')
plt.title('Title Goes here')
plt.show()
```

```python
x2 = [1, 2, 3]
y2 = [11, 12, 15]
plt.plot(x2, y2, label = 'Primeira Linha')
plt.legend()
plt.show()
```

## Gráfico de Barras

```python
x = [2, 4, 6, 8, 10, 12]
y = [6, 7, 8, 2, 4, 5]
plt.bar(x, y, label = 'Barras', color = 'black')
plt.legend()
plt.show()
```

```python
idades = [22, 65, 45, 55, 21, 22, 23, 34, 42, 10, 11, 9, 2, 8, 80, 100, 81, 91, 90]
index = [x for x in range(len(idades))]
plt.bar(index, idades, label = 'Idades', color = 'gray')
plt.show()
```

```python
x2 = [1, 3, 5, 7, 9, 13]
y2 = [5, 6, 7, 12, 10, 6]
plt.bar(x, y, label = 'Barras', color = 'red')
plt.bar(x2, y2, label = 'Barras 2', color = 'yellow')
plt.legend()
plt.show()
```

```python
# Criando histogramas
bin = [i * 10 for i in range(14)]
```

```python
plt.hist(idades, bin, histtype = 'bar', rwidth = 0.8, color = 'green')
```

```python
plt.hist(idades, bin, histtype = 'stepfilled', rwidth = 0.8, color = 'green')
```

# ScatterPlot

```python
x = [x + 1 for x in range(8)]
y = [5, 2, 4, 5, 6, 8, 4, 8]
```

```python
# marker = 'o' será de pontos
plt.scatter(x, y, label = 'ScatterPlot', color = 'red', marker = 'o', s = 100)
```

```python
# marker = 'x' será de X's
plt.scatter(x, y, label = 'ScatterPlot', color = 'red', marker = 'x', s = 100)
```

# StackPlot

```python
dias = [x + 1 for x in range(5)]
dormir = [7, 8, 6, 77, 7]
comer = [2, 3, 4, 5, 3]
trabalhar = [7, 8, 7, 2, 2]
passear = [8, 5, 7, 8, 13]
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors = ['yellow', 'cyan', 'red', 'black', 'blue'])
plt.show()
```

# Pie Chart

```python
slices = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'trabalhar', 'passear']
colunas = ['c', 'm', 'r', 'k']
plt.pie(slices, labels = atividades, colors = colunas, startangle = 90, shadow = True, explode = (0, 0.1, 0, 0))
plt.show()
```
