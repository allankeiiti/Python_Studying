# Plotting

```python
from matplotlib import style
%matplotlib inline
```

```python
# Time Series Plot
timeSeries = pd.Series(np.random.randn(500), index = pd.date_range('1/1/2016', periods = 500))
timeSeries = timeSeries.cumsum()
timeSeries.plot()
```

```python
# DataFrame Plot
dataFrame = pd.DataFrame(np.random.randn(500, 4), index = timeSeries.index, columns = ['A', 'B', 'C', 'D'])
dataFrame = dataFrame.cumsum()
plt.figure(); dataFrame.plot(); plt.legend(loc = 'best')
```
